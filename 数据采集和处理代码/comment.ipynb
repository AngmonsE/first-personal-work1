```python
import re
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36'
}
cursor='0'
_='1614279087875'
comment_list=[]
```

```python
for i in range(2000):
    url='https://coral.qq.com/article/5963120294/comment/v2?callback=_article5963120294commentv2&orinum=10&oriorder=o&pageflag=1&cursor='+cursor+'&scorecursor=0&orirepnum=2&reporder=o&reppageflag=1&source=1&_='+_
   
    html=requests.get(url,headers=headers).content.decode()
   
    comment=re.findall('"content":"(.*?)","',html, re.S)
    
    comment_list.append(comment)
    
    cursor=re.findall('"last":"(.*?)","',html, re.S)[0]
```

```python
comment_list
```

```python
with open("comment.txt", "w", encoding="utf-8") as f:
    for i in comment_list:
        for j in i:
            f.write(j)
            f.write("\n")
```
